```python
# Fungsi untuk menghitung efisiensi turbin
def hitung_efisiensi(input_energi, output_energi):
    efisiensi = output_energi / input_energi
    return efisiensi

# Contoh pemanggilan fungsi (bonus)
hasil = hitung_efisiensi(500, 400)
print("Efisiensi turbin =", hasil)
```

Penjelasan Program No 2.
Fungsi hitung_efisiensi menerima dua parameter: input_energi dan output_energi.
Rumus efisiensi: Output / Input.
Nilai hasil perhitungan dikembalikan dengan return.
Pemanggilan fungsi dilakukan dengan nilai input_energi = 500 dan output_energi = 400.

```python
# Program menghitung daya listrik generator

# Diketahui
tegangan = 220   # dalam Volt
arus = 15        # dalam Ampere

# Rumus daya listrik
daya = tegangan * arus

# Tampilkan hasil
print("Tegangan (V):", tegangan, "Volt")
print("Arus (I):", arus, "Ampere")
print("Daya listrik (P):", daya, "Watt")
```

```python
# Fungsi/Modularisasi
def hitung_efisiensi_termal():
    energi_masuk = float(input("Masukkan energi masuk (Joule): "))
    energi_keluar = float(input("Masukkan energi keluar (Joule): "))

    efisiensi = (energi_keluar / energi_masuk) * 100

    print(f"Efisiensi termal turbin gas = {efisiensi:.2f}%")

hitung_efisiensi_termal()
```

```python
# While loop & Logika
tangki = 0

while tangki < 100:
    tangki += 10
    print(f"Tangki terisi {tangki} liter")

    if tangki == 80:
        print("WARNING: Hampir penuh!")
    elif tangki == 100:
        print("TANGKI PENUH ‚Äì Pengisian berhenti.")

```

Perulangan akan terus berjalan selama isi tangki kurang dari 100 liter. Artinya, program akan berhenti secara otomatis ketika tangki sudah mencapai atau melebihi 100 liter.
