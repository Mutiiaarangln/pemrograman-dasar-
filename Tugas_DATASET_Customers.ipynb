```python
# 1. Import Library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 2. Load Dataset
df = pd.read_csv("customers.csv")

print(" 5 Data Teratas ")
print(df.head())
```

```python
# 3. Informasi Dataset
print("\n Info Dataset ")
print(df.info())
print("\n Deskripsi Statistik ")
print(df.describe())
```

```python
# 4. Mengecek Missing Value
print("\n Cek Missing Value ")
print(df.isnull().sum())
```

```python
# 5. Mengecek Duplikasi
print("\n Cek Data Duplikat ")
print("Jumlah duplikat:", df.duplicated().sum())
```

```python
# 6. Analisis Kolom Gender
print("\n Jumlah Gender ")
print(df['gender'].value_counts())

# Visualisasi
plt.figure()
df['gender'].value_counts().plot(kind='bar')
plt.title("Distribusi Gender")
plt.xlabel("Gender")
plt.ylabel("Jumlah")
plt.show()
```

```python
# 7. Analisis Kolom Age
plt.figure()
df['age'].plot(kind='hist')
plt.title("Distribusi Usia Pelanggan")
plt.xlabel("Usia")
plt.ylabel("Frekuensi")
plt.show()

print("\n Statistik Usia ")
print(df['age'].describe())
```

```python
# 8. Analisis Berdasarkan State
print("\n Jumlah Pelanggan per State ")
print(df['state'].value_counts())

plt.figure()
df['state'].value_counts().plot(kind='bar')
plt.title("Jumlah Pelanggan per State")
plt.xlabel("State")
plt.ylabel("Jumlah")
plt.show()
```
